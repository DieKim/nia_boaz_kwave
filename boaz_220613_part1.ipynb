```python
###연상이미지_for_top10
#여러 가지 후보 중(글로 설명) top10 보여주기
#지금 너무 뒤죽박죽이라 형식 통일 필요

연상 = pd.read_csv('BOAZXNIA/REMIND_IMAGE_INFO_2021(연상이미지).csv')
연상['IMAGE_NM'] = 연상['IMAGE_NM'].apply(lambda x: x.replace(" ", ""))
print(연상['IMAGE_NM'].unique())

def for_image(x):
  if x in ['K-Pop', 'K-POP']:
    return 'K-POP'
  elif x in ['한국음식(김치,김,볼고기등)', '한국음식(김치,김,불고기등)', '한국음식(김치,불고기등)', '한식']:
    return '한식'
  elif x in ['드라마', '영화', '예능프로고램', '예능프로그램']:
    return '영화/드라마/예능'
  elif x in ['한류스타(BTS,이민호등)', '한류스타(이민호,싸이등)', '연예인']:
    return '한류스타'
  elif x in ['태권도', '스포츠/스포츠스타', '올림픽/월드컵', '올림픽1월드컵', '스포즈/스포즈스타', '스포츠스타(김연아,박지성등)', '스포츠스타']:
    return '스포츠'
  elif x in ['한국전쟁(TheKoreanWarin1950)', '북핵위협/전쟁위험지역/북한', '북한', 
             '북핵위협/전쟁위험지역', '북한/북핵위협/전쟁위험', '북한/북핵위협/전쟁위험지역',
             '북한(분단국가)' '한국전쟁']:
    return '북한/전쟁'
  elif x in ['IT제품/브랜드(휴대폰TV,컴퓨터등)', 'IT제품/브랜드(휴대폰,TV,컴퓨터등)', 'IT산업(휴대폰,TV,컴퓨터등)', 'IT첨단산업']:
    return 'IT'
  elif x in ['온라인/모바일게임/e-sports/프로게이머', '온라인/모바일게임', '게임']:
    return '게임'
  elif x in ['한글(한국어)', '한복/한옥/궁궐', '한복', '한옥/궁궐(경복궁,광화문,등)']:
    return '문화유산' 
  elif x in ['관광지(서울,제주,남산타워등)', '관광지']:
    return '관광'
  elif x in ['만화/웹툰/도서/문학', '애니메이션', '만화', '도서']:
    return '애니메이션/만화/웹툰/문학'
  elif x in ['뷰티서비스(성형)', '뷰티제품(화장품)', '패션', '패선', '뷰티제품', '뷰티(화장품및성형)', '미용']:
    return '뷰티/패션'
  elif x in ['경제성장', '새마을운동']:
    return '경제'
  else:
    return '기타'

연상['IMAGE_NM'] = 연상['IMAGE_NM'].apply(for_image)
print(연상['IMAGE_NM'].unique())
```

```python
세종 = pd.read_csv('BOAZXNIA/KC_SEJONGKOREALANGAUGEINSTITUTE_2020(세종학당).csv')
세종
```

```python
#칼럼 수정
세종 = 세종.iloc[:, :-2]
세종 = 세종.rename(columns={'SURVEY_BASE_DATE_YY': '조사연도',
                         'TOTAL_NATION_CO': '총 국가 수',
                         'TOTAL_PLACE_CO': '총 장소 수',
                         'STUDENT_CO': '학생 수',
                         'GRADUATE_CO': '졸업생 수',
                         'ONLINE_STUDENT_CO': '온라인 학생 수',
                        'ONLINE_CLASS_CO': '온라인 수업 수'
                        })
세종
```

```python
#wide -> long by melt
세종 = 세종.melt(id_vars='조사연도')
세종
```

```python
세종.to_csv('세종학당.csv', encoding='utf-8-sig', index=False)
```

```python
###문화콘텐츠소비비중 전처리
def for_content_spend(x):
  if x in ['소비 비중(%)', '소비비중(%)', '소비비중 (%)']:
    return '소비비중(%)'
  elif x in ['한국 문화콘텐츠 소비(3개월,편)', '한국 문화콘텐츠 소비(편)']:
    return '한국 문화콘텐츠 소비량(편)'
  elif x in ['한국 문화콘텐츠 소비(시간)', '한류콘텐츠소비량(시간)']:
    return '한국문화콘텐츠 소비량(시간)'
  elif x in ['한국 문화콘텐츠 소비($)', '한류콘텐츠소비량($)']:
    return '한국 문화콘텐츠 소비량($)'
  elif x in ['월평균 지출 금액($)', '월평균 지출 금액(3개월,$)' ]:
    return '월평균 지출 금액($)'
  elif x in ['전체 소비량(시간)', '전체소비량(시간)']:
    return '전체 소비량(시간)'
  elif x in ['전체 소비량(편)']:
    return '전체 소비량(편)' 
  elif x in ['전체 소비량($)', '전체소비량($)']:
    return '전체 소비량($)' 

소비 = pd.read_csv('BOAZXNIA/CLTUR_CNTNTS_CNSMP_INFO_2021(문화컨텐츠소비비중).csv')
소비['IEM_NM'] = 소비['IEM_NM'].apply(for_content_spend)
소비 = 소비[['IEM_NM', 'MALE_RATE', 'FEMALE_RATE']]
소비 = 소비.melt(id_vars='IEM_NM')
소비.to_csv('문화콘텐츠소비비중_donut2.csv', encoding='utf-8-sig', index=False)
```
